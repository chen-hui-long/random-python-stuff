```python
# import libraries 
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# dataset 
diamonds = sns.load_dataset("diamonds")
diamonds.head()
```

```python
pd.crosstab(diamonds.color, diamonds.cut).head() # 'color' as the index and 'cut' s the columns 
```

```python
pd.crosstab(diamonds.color, diamonds.cut, values = diamonds.price, aggfunc = np.mean).fillna(0).round().head()
```

```python
_ = pd.crosstab(diamonds.color, diamonds.cut, values = diamonds.price, aggfunc = np.mean).fillna(0).round()
sns.heatmap(_, annot =True, fmt = 'g')
```
